```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return "\n".join(text)
```

```python
import re
from typing import Dict, Any

def parse_habr_article(content: str) -> Dict[str, Any]:
    result = {
        "company_name": None,
        "author": None,
        "company_description": None,
        "rating": None,
        "publication_date": None,
        "article_text": []
    }

    # Поиск заголовка статьи
    title_match = re.search(r'# (.*?)\n', content)
    if title_match:
        result["article_text"].append(title_match.group(1).strip())

    # Поиск информации об авторе/компании
    company_header = re.search(r'(.+?)\n+([^\n]+?)\n+(\d+\.\d+)\n+Рейтинг', content)
    if company_header:
        result["company_name"] = company_header.group(1).strip()
        result["company_description"] = company_header.group(2).strip()
        result["rating"] = float(company_header.group(3))
    else:
        user_match = re.search(r'\*\*([^\n]+?)\n+\d+ мин\s+(\d+)\s+', content)
        if user_match:
            result["author"] = user_match.group(1).strip()
            result["rating"] = int(user_match.group(2))

    # Дата публикации
    date_match = re.search(r'(\d+ (?:минут|час|день|дня|дней|месяц|года) назад)', content)
    if date_match:
        result["publication_date"] = date_match.group(1)

    # Основной текст статьи (эвристический подход)
    text_sections = re.split(r'===== Page \d+ =====|Теги:|Хабы:', content)
    if len(text_sections) > 1:
        clean_text = []
        for line in text_sections[1].split('\n'):
            line = line.strip()
            if line and not re.match(r'^\*|→|←|©|\d+\.\d+|\d+ мин|\d+ \d+', line):
                clean_text.append(line)
        result["article_text"] += [p for p in clean_text if p]
    
    return result
```
