## Парсинг топ 250 фильмов IMDb

<strong>Задача</strong>. При помощи библиотек для прасинга (BeautifulSoup, selenium) провести парсинг данных о топе 250 фильмов по версии IMDb с оффициального сайта IMDb. В результирующую выборку обязательно должны входить: название фильма, его описание, год выпуска, длительность показа и рейтинг. Дополнительно будет парситься возрастное ограничение.  

После парсинга привести данные в порядок: поработать с пропущенными значениями, возможными дубликатами и типами данных.

### 1.1 Импорт модулей

```python
# Для работы с данными
import pandas as pd

# Для визуализации
import seaborn as sns
import matplotlib.pyplot as plt

# Для парсинга
from bs4 import BeautifulSoup as bs
from selenium import webdriver

# Для получения html кода сайта
import requests

# Для перевода с ангийского на русский
from googletrans import Translator

# Для предобработки данных
from sklearn.preprocessing import LabelEncoder

# Отключение уведомлений
import warnings
warnings.filterwarnings('ignore')
```

### 1.2 Парсинг

В начале парсинга попробуем сделать запрос на сайт IMDb, чтобы понять, возможен ли парсинг сайта

```python
# Начинаем сессию
req = requests.Session()

# Определяем заголовки, чтобы при запросе имитировать пользователя
user = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36'
st_accept = "text/html"
headers = {"Accept":st_accept, "User-Agent": user}

# Определяем ссылку и пытаемся сделать запрос
url = 'https://www.imdb.com/chart/top/'
response = req.get(url, headers=headers)

# Выводим статус код
response.status_code
```

Статус код - 200, а значит, что сайт корректно ответил на запрос и вернул HTML код страницы

Определим словарь, в который будем записывать спаршенные данные, чтобы потом преобразовать в датасет.  
Как уже говорилось в постановке задачи, результирующая выборка будет содержать: название фильма, его описание, год выпуска, длительность показа, рейтинг и возрастное ограничение.

```python
films = {'title':[], 'description':[], 'year':[], 'duration':[], 'age_limit':[], 'rating':[]}
```

Ну а теперь сам парсинг. Будет он происходить следующим образом: 
1. Получаем html-код всего топа 250 сразу
2. Определим сипсок в который спарсим ссылки на все 250 фильмов
3. Проходимся по всему списку, переходя на каждую страничку с информацией о каждом фильме
4. Парсим необходимые данные каждого фильма
5. Добавляем данные в словарь

```python
# Делаем запрос при помощи selinium, т.к. он может прогрузить всю страничку сразу
url = 'https://www.imdb.com/chart/top/'
driver = webdriver.Chrome()
driver.get(url)
response = driver.page_source

# Вытаскиваем html-код из результата запроса
soup = bs(response, 'html.parser')

# Определяем список, в который спарсим все ссылки на странички фильмов
titles = soup.find_all('a', class_='ipc-title-link-wrapper', limit=250)
# Проверим, точно ли фильмов 250 штук
print(len(titles))

# Пройдёмся по каждому фильму в частности
index = 1
for title in titles:
    # Вытаскиваем непосредственно ссылку
    film_href = title.get('href')
    
    # Делаем запрос и получаем html-код странички фильма
    url = 'https://www.imdb.com' + film_href
    response = req.get(url, headers=headers)
    soup = bs(response.text, 'html.parser')

    '''
    film_title = soup.find('div', class_='sc-ec65ba05-1 fUCCIx')
    if film_title is not None:
        film_title = film_title.text.split(': ')[1]
    else:
    '''

    # Для начала спарсим название 
    film_title = soup.find('span', class_='hero__primary-text').text

    # После парсим рейтинг
    film_rating = soup.find('span', class_='sc-d541859f-1 imUuxf').text

    # Далее парсим описание фильма
    film_desc = soup.find('span', class_='sc-42125d72-0 gKbnVu').text
    # Так как описание у фильмов на английском языке, при помощи модуля googletrans переведём описание фильма
    async with Translator() as translator:
        film_desc = await translator.translate(film_desc, dest = 'ru')

    # Наконец парсим прочую информацию о фильме, которая содержит в себе год выпуска, длительность показа и возрастное ограничение
    film_info = soup.find('ul', class_='ipc-inline-list ipc-inline-list--show-dividers sc-ec65ba05-2 joVhBE baseAlt')

    # Так как не на всех фильмах указано возрастное ограничение, ставим условие
    if len(list(film_info.stripped_strings)) == 3:
        # Если есть все пункты, то записываем всё сразу в 3 переменные
        film_year, film_age_limit, film_duration = list(film_info.stripped_strings)
    else:
        # Если же позиций меньше, чем 3, то в таком случае нам не хватает возрастного ограничения
        film_year, film_duration = list(film_info.stripped_strings)
        # Его мы ставим в ручную
        film_age_limit = 'Not Rated'

    # Для проверки выводим инфомрацию о фильме
    print(f'{index}.')
    print(f'Название: {film_title} {film_rating}\nОписание: {film_desc.text}\n')
    print(f'Год: {film_year}\nВозрастное ограничение: {film_age_limit}\nДлительность: {film_duration}\n\n')

    # Наконец, добавляем всю информацию о фильме в словарь
    films['title'].append(film_title)
    films['description'].append(film_desc.text)
    films['year'].append(film_year)
    films['age_limit'].append(film_age_limit)
    films['duration'].append(film_duration)
    films['rating'].append(film_rating)
    
    index += 1
```

### 1.4 Предобработка данных

Чтобы проводить дальнейший анализ, нужно немного поработать с датасетом. Поработать в том плане, что поменять типы данных для некоторых признаков, а некоторые и то удалить из датасета.

Под удалением, а скорее даже заменой, имеется ввиду преобразование столбца с длительностью фильма. Так как длительность на IMDb указана в неудобном для анализа формате, было бы намного удобнее работать, если столбец, например, был бы разделён на часы и минуты отдельно. Так и тип данных у столбцов будет соответствующим, так ещё и простор для анализа будет больше.  

Чтобы достичь поставленной цели, напишем функцию, которая будет возвращать часы и минуты длительности по отдельности, даже если указаны только минуты либо часы.

```python
def convert_duration(duration):
    # Сначала рассмотрим случай, когда указаны часы и минуты сразу
    if len(duration.split(' ')) == 2:
        # В таком случае сплитуем длительнось так, чтобы часы и минуты были отдельными элементами списка
        duration = duration.split(' ')
        # Теперь часы сплитуем по букве h, так мы отделим кол-во часов
        hours = duration[0].split('h')[0]
        # Тоже самое проделываем с минутами
        minutes = duration[1].split('m')[0]

    # Теперь рассмотрим случай, когда указаны только часы
    elif len(duration.split(' ')) == 1 and 'h' in duration:
        # В таком случае махинации по получению часов идентичны прошлому ветвлению
        hours = duration.split('h')[0]
        # А минуты ставим на 0
        minutes = 0

    # Последний случай, когда указаны только минуты
    else:
        # Тут всё аналагично прошлому elif
        minutes = duration.split('m')[0]
        hours = 0

    return hours, minutes
```

Функция написана, а значит пора её применить.  
Помимо этого сделаем ещё несколько преобразований:
- изменим тип данных признака 'years' на int64
- изменим тип данных признака 'rating' на float64

```python
# Смена типов данных для years и rating
df['year'] = df['year'].astype('int64')
df['rating'] = df['rating'].astype('float64')

# Создаём новые признаки, применяя ранее написанную функцию к столбцу duration
df['minutes'] = df['duration'].apply(lambda x: convert_duration(x.strip())[1])
df['hours'] = df['duration'].apply(lambda x: convert_duration(x.strip())[0])

# Сразу меняем тип данных на int64 
df['minutes'] = df['minutes'].astype('int64')
df['hours'] = df['hours'].astype('int64')
```
