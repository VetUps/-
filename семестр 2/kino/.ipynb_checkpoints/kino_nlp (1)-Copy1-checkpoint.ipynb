### Предобработка текста

Для начала загружаем данные, которые ранее были спаршены с сайта IMDB

```python
df = pd.read_csv('kino_nlp.csv', index_col=[0])
df.head(10)
```

Хорошо, теперь взглянем на описание первого фильма, чтобы понять с чем будет вестись работа

Чтобы подготовить такой текст к тематическому моделированию, необходимо провести все базовые преобразования:  
- избавиться от пунктуации и стоп слов
- токенизировать текст
- лемматизировать текст

```python
# Функция для удаления знаков пунктуации
def remove_punctuation(text):
    punctuation = string.punctuation
    return ''.join([ch for ch in text if ch not in punctuation])

# Функция для удаления стоп слов
def remove_stopwords(tokenised_text):
    stop_words = stopwords.words('russian')
    stop_words.extend(['свой', 'год', 'который'])
    clean_tokenised_text = [word for word in tokenised_text if word not in stop_words]
    return clean_tokenised_text

# Функция для токенезации
def tokenise(text):
    words = nltk.word_tokenize(text)
    return words

# Функция для лемматизации
def lemmatizing(tokenised_text):
    morph = pymorphy3.MorphAnalyzer(lang='ru')
    lemmtizing_text = [(morph.parse(word)[0]).normal_form for word in tokenised_text]
    return lemmtizing_text
```

Все функции подготовлены. Первым делом токенизируем текст попутно удаляя пунктуацию и стоп-слова

```python
df['clean_tokenised_description']= [remove_stopwords(tokenise(remove_punctuation(text.lower()))) for text in df['description']]
df.head(5)
```

Теперь текст разбит на токены, но слова все ещё не преведены к лемме, надо это исправить

```python
df['lemmatized_description']= [remove_stopwords(lemmatizing(tokenised_text)) for tokenised_text in df['clean_tokenised_description']]
df.head(5)
```

Отлично, теперь текст преведён к такой форме, в какой можно провести его анализ и в последующем тематическое моделирование

### Тематическое моделирование

```python
# Модели тематического моделирования
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.decomposition import NMF
from sklearn.decomposition import TruncatedSVD
from bertopic import BERTopic

# Модели для определения количества тематик
from gensim.models import CoherenceModel, LdaModel
# Прочие вспомогательные библиотеки
import gensim.corpora as corpora
```

Чтобы обучить модели тематического моделирования, необходимо определить количество тематик. Для этого обучим модель LDA с количеством тематик от 3 до 11 и для каждой обученной модели посмотрим когерентность.  
  
Для модели необходим немного другой мешок слов вида id_слова:частота

```python
texts = [text.split() for text in df['lemmatized_description']]

# Составляем словарь уникальных слов
gensim_dictionary = corpora.Dictionary(texts)
# Составляем мешок слов
corp = [gensim_dictionary.doc2bow(text) for text in texts]
```

```python
corp
```

Когда мешок слов образован, прообучаем модель несколько раз и посмотрим на результатыт

```python
# Список для записи когерентностей моделей
gensim_coherence_values = []
# Список для записи моделей
model_list = []

# В цикле переберём количество тем от 3 до 11
for num_topics in range(3, 20):
    # Инициализируем модель, указывая корпус и количество тем
    model = LdaModel(corpus=corp, 
                     num_topics=num_topics, 
                     random_state=42)

    # Добавляем модель в список
    model_list.append(model)

    texts = [text.split() for text in df['lemmatized_description']]
    # Рассчитыаем когерентность модели
    coherencemodel = CoherenceModel(model=model, 
                                    texts=texts, 
                                    dictionary=gensim_dictionary, 
                                    coherence="c_v")

    # Записываем результат в список
    gensim_coherence_values.append(coherencemodel.get_coherence())

# Выводим результаты
for i in range(len(model_list)):
    print(f'Model: {i} Topics: {i+3} Coherence: {gensim_coherence_values[i]}')
```

Видим, что до 7 тематик когерентость растёт, а затем падает и растёт снова, почти до прежнего значения. В итоге получается, что при количестве тематик равным 7, достигается максимальная когерентность. Значит возьмём 7 тематик для обучения последующих моделей

```python
num_topics = 11
```

Для составления тем, будем использовать следующие модели:  
  
**Latent Dirichlet Allocation (LDA):**

Вероятностная модель, которая предполагает, что каждый документ является смесью нескольких тем, а каждая тема — это распределение вероятностей над словами.

Пример: Если в документах часто встречаются слова "кошка", "собака", "попугай", модель может выделить тему "домашние животные".

**Non-negative Matrix Factorization (NMF):**

Матричная факторизация, которая разлагает матрицу "термин-документ" на две неотрицательные матрицы: темы и их веса в документах.

Пример: Если в документах часто встречаются слова "футбол", "мяч", "гол", модель может выделить тему "спорт".

**Latent Semantic Analysis (LSA):**

Метод, основанный на сингулярном разложении (SVD) матрицы "термин-документ".

Пример: Если в документах часто встречаются слова "компьютер", "программа", "алгоритм", модель может выделить тему "информационные технологии".

#### LDA

Для начала получим тематики при помощи модели LDA

```python
# Создание модели LDA
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(X_counts)

# Вывод топ слов для каждой темы
for topic_idx, topic_words in enumerate(lda.components_):
    # Получаем индексы самых весомых слов
    top_words_idx = topic_words.argsort()[-10:][::-1]
    # По индексам получаем сами слова их векторизатора
    top_words = [vectorizer.get_feature_names_out()[i] for i in top_words_idx]
    # Выводим темы
    print(f"Тема {topic_idx + 1}: {', '.join(top_words)}")
```

Получилось 7 тем, в выводе показаны слова, которые имеют наибольший вес в данной теме. Перейдём к следующей модели 

#### NMF

Далее идёт модель NMF

```python
# Создание модели NMF
nmf = NMF(n_components=num_topics, random_state=42)
nmf.fit(X_counts)

# Вывод топ слов для каждой темы
for topic_idx, topic_words in enumerate(nmf.components_):
    # Получаем индексы самых весомых слов
    top_words_idx = topic_words.argsort()[-10:][::-1]
    # По индексам получаем сами слова их векторизатора
    top_words = [vectorizer.get_feature_names_out()[i] for i in top_words_idx]
    # Выводим темы
    print(f"Тема {topic_idx + 1}: {', '.join(top_words)}")
```

Также опредили 7 тем и складывается такое, чувство, что 4 тема явно связана с фильмами про Звёздные Войны

#### LSA

Далее тематики сгенерирует LSA

```python
# Создание модели LSA
lsa = TruncatedSVD(n_components=num_topics, random_state=42)
lsa.fit(X_counts)

# вывод топ слов для каждой темы
for topic_idx, topic_words in enumerate(lsa.components_):
    # Получаем индексы самых весомых слов
    top_words_idx = topic_words.argsort()[-10:][::-1]
    # По индексам получаем сами слова их векторизатора
    top_words = [vectorizer.get_feature_names_out()[i] for i in top_words_idx]
    # Выводим темы
    print(f"Тема {topic_idx + 1}: {', '.join(top_words)}")
```

Тут также 7 тем, где просматриваются базовые темы про войну и Звёздные войны

### Классификация

```python
# Модели классификации
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from catboost import CatBoostClassifier

# Метрики класификации
from sklearn.metrics import f1_score, classification_report, make_scorer, confusion_matrix
# Метод кросс-валидации и разделения набора данных
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
y = df['topic']

# Правильный способ получить эмбеддинги через BERTopic
embeddings = topic_model._extract_embeddings(
    df['lemmatized_description'].tolist(),  # Тексты в виде списка
    method="word",             # Использует встроенный эмбеддинг-модель
    verbose=True
)
X = embeddings

# Разделяем выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
tf_idf_vectorizer = TfidfVectorizer(stop_words=stopwords.words('russian'),
                                    max_df=0.8, min_df=0.02,
                                    ngram_range=(1, 3))

X_train = tf_idf_vectorizer.fit_transform(X_train)
X_test = tf_idf_vectorizer.transform(X_test)
```

```python
def estimating_model(model):
    # Тренируем модель и делаем предсказание
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    # Замеряем f1_score
    f1 = f1_score(y_test, pred, average='weighted')
    print(f'F1 score - {f1}')

    # Выводим classification_report
    print(classification_report(y_test, pred))

    # Возвращаем метрики и предсказание
    return f1, pred
```

```python
knn = KNeighborsClassifier(weights='distance')
knn_f1, knn_pred = estimating_model(knn)
```

```python
gbc = GradientBoostingClassifier()
gbc_f1, gbc_pred = estimating_model(gbc)
```

```python
cb = CatBoostClassifier(logging_level='Silent', auto_class_weights='Balanced')
cb_f1, cb_pred = estimating_model(cb)
```
